# Reshaping , stacking , squeezing and unsqueezing 

- view - return a view of a tensor but keeps the memory of tensor same
- hstack , vstack
- squeeze - remove one dimension 
- unsqueeze - add on dimension 
- permute - return a view of input with dimensions permuted(swapped)

```python
x = torch.arange(1.,10.)
x , x.shape
```

```python
x_reshape = x.reshape(1,9)
x_reshape, x_reshape.shape
```

```python
x_reshape = x.reshape(9, 1)
x_reshape, x_reshape.shape
```

```python
# change view 
z = x.view(1,9)
z , z.shape
```

```python
# changing z changes x 
z[:,0] = 5 
z,x 
```

```python
# stack tensors 
x_stack = torch.stack([x,x,x,x],dim=0) # hstack
x_stack
```

```python
# stack tensors
x_stack = torch.stack([x, x, x, x], dim=1) # vstack
x_stack
```

```python
# torch.squeeze() - removes all single dimensions from target tensor 
x_reshape.shape
```

```python
x_reshape.squeeze()
```

```python
x_reshape.squeeze().shape
```

```python
print(f"Previous tensor: {x_reshape}")
print(f"Previous shape: {x_reshape.shape}")

print()
x_squeeze = x_reshape.squeeze()
# removes extra dimensions
print(f"New tensor: {x_squeeze}")
print(f"New shape: {x_squeeze.shape}")
```

```python
# torch.squeeze() - add a single dimension to target tensor at specific dim
print(f"Previous tensor: {x_squeeze}")
print(f"Previous shape: {x_squeeze.shape}")

print()
x_unsqueeze = x_squeeze.unsqueeze(dim=0)
# removes extra dimensions
print(f"New tensor: {x_unsqueeze}")
print(f"New shape: {x_unsqueeze.shape}")
```

```python
# torch.permute() - rearranges dimension of a target tensor in specific order
x_original = torch.rand(size=(244,244,3)) # h w c 
print(f"Previous shape: {x_original.shape}")

# permute the original tensor to rearrange axis order
x_permute = x_original.permute(2,0,1) # shifts to 0->1 , 1->2 , 2->0
print(f"New shape: {x_permute.shape}")

# same memory just show differently
```

# Reproducibility (taking out random from random)

to reduce randomness , use random seed. (pseudo random)
random seed fixes the randomness  

```python
# random but reproducable tensors

RANDOM_SEED = 42 
torch.manual_seed(RANDOM_SEED)
# matrix 1 :
random_a = torch.rand(3,4)

# have to do it again as requesting next random will not be same
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
# matrix 2 :
random_b = torch.rand(3, 4)

print(random_a == random_b)
```
